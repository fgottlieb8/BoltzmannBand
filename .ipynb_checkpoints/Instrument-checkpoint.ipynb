```python
from midiutil.MidiFile import MIDIFile
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
#map s into energy. s=1 -> minor 2nd, s=2 major 2nd, s=3 minor 3rd, s=4 major 3rd, s=5 fourth, s=6 tritone, s=7 5th
def sPitchMap():
    i0=1
    im2=16*15 #240
    i2=9*8    #72
    im3=6*5   #30
    i3=5*4    #20
    i4=4*3    #12
    it=64*45  #2880
    #it=500
    i5=3*2    #6
    im6=8*5   #40
    i6=5*3    #15
    im7=16*9  #144
    i7=15*8   #120
    return [i0,im2,i2,im3,i3,i4,it,i5,im6,i6,im7,i7]

sPitch=sPitchMap()
sPitch
```

```python
#Generate a dictionary that matches pitches to energy
#pitches go from C0=0 G9=127

def genEnergyMap(bp=basepitch):
    test=[]
    octave_feature=8
    for p in range(128):
        #print(p)
        s=(p-bp)%12
        m=abs(p-bp-s)/12
        energy=octave_feature*m+sPitch[s]
        test.append(energy)
    return np.array(test)
    
```

```python
#energyMap is an np.array that maps each pitch to an energy
# energyMap=genEnergyMap()

def boltzmanDistributionRaw(energyMap,T):
    return np.exp(-energyMap/T)

def partitionFunction(energyMap,T):
    allFactors=boltzmanDistributionRaw(energyMap,T)
    Z=np.sum(allFactors)
    return Z
    
def boltzmanDistribution(energyMap,T):
    Z=partitionFunction(energyMap,T)
    return np.divide(np.exp(-energyMap/T),Z)


plt.plot(boltzmanDistribution(genEnergyMap(),5))
```

```python
#Sample n Notes from harmony distribution
def samplePitch(n,energyMap,T):
    distribution=boltzmanDistribution(energyMap,T)
    return np.random.choice(np.arange(len(distribution)),n,p=distribution)
```

```python
#Starting Rythem Section
#What note lengths do I want to allow? Nothing longer than a measure. 
#Start with m/2^n for n goes to 2^6

def genRythemEnergyMap(): 
    n=6
    result=np.full(2**n,2**n)
    for i in range(n-1,-1,-1): #goes from 2,1,0
        for j in range(1,2**i+1):  # goes 4,2,1
            result[2**(n-i)*j-1]=i**1.5 #changed 2 to 1.5
    return result

def sampleRythem(n,energyMap,T): #returns amount of quarter note time
    distribution=boltzmanDistribution(energyMap,T)
    return np.divide(np.add(np.random.choice(np.arange(len(distribution)),n,p=distribution),1),16)

sampleRythem(3,genRythemEnergyMap(),T=6)
```

```python
#Making chord song
#MIDIFile.addNote(track, channel, pitch, time, duration, volume, annotation=None)
def makeChordSong(T=5, songlength=128, tempo=120):
    song=MIDIFile()
    song.addTempo(0,0,tempo)
    time=0
    pEM=genEnergyMap()
    rEM=genRythemEnergyMap()
    while time<songlength:
        pitches=samplePitch(3,pEM, T)
        lengths=sampleRythem(1,rEM,T)
        song.addNote(0,0,pitches[0],time, duration=lengths[0],volume=100)
        song.addNote(0,0,pitches[1],time, duration=lengths[0],volume=100)
        song.addNote(0,0,pitches[2],time, duration=lengths[0],volume=100)
        time+=lengths[0]
    return song
```

```python
#making multi channel song
def makeSong(T,n=5,songlength=128):
    song=MIDIFile(numTracks=n)
    if type(T) is int:
        T=np.full(songlength,T)
    if len(T)>songlength:
        T=T[0:songlength]
    pEM=genEnergyMap()
    rEM=genRythemEnergyMap()
    for track in range(n):
        time=0
        while time<songlength:
            pitch=samplePitch(1,pEM, T[math.floor(time)])[0]
            length=sampleRythem(1,rEM,T[math.floor(time)])[0]
            song.addNote(track,0,pitch,time, duration=length,volume=100)
            time+=length
    return song
    
    
```
